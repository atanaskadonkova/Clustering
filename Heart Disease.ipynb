```python
# C) Logistic Regression:
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# a. Build a simple logistic regression model

X = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
y = df['target']



```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# i. Split the dataset into 70:30 ratio
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# ii. Build the logistic regression model on the training set 
logreg = LogisticRegression()
logreg.fit(X_train, y_train)


```

```python
# predict the values on the test set

y_pred = logreg.predict(x_test)
```

```python
# iii. Build the confusion matrix and get the accuracy score

from sklearn.metrics import accuracy_score
accuracy_score (y_train, logreg.predict(X_train))
```

```python
accuracy_score (y_test, logreg.predict(x_test))
```

```python
# iii. Build the confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Get the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy_score:", accuracy)
```

```python
# E) Random Forest:
```

```python
# i. Divide the dataset in 70:30 ratio

X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30)  
```

```python
# ii. Build the model on train set and predict the values on test set
```

```python
# iii. Build the confusion matrix and calculate the accuracy
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)

```

```python
y_pred=clf.predict(x_test)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# iv. Visualize the model using the graphviz package
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz
import graphviz



# Get an individual decision tree from the random forest
tree = clf.estimators_[0]  # Change the index to select a different tree

# Export the decision tree to DOT format
dot_data = export_graphviz(tree, out_file=None, feature_names=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal'])
graph = graphviz.Source(dot_data)
graph.view()
```

```python
# F) Select the best model
```

```python
# a. Print the confusion matrix of all classifiers

classifiers = [logreg, clf, tree]  # List of trained classifiers
test_data = X_test  # Test data used for evaluation
test_labels = y_test  # True labels for the test data

for clf in classifiers:
    y_pred = clf.predict(test_data)
    cm = confusion_matrix(test_labels, y_pred)
    print("Confusion Matrix for", clf.__class__.__name__)
    print(cm)
    print()
```

```python
# b. Print the classification report of all classifiers

from sklearn.metrics import classification_report

for clf in classifiers:
    y_pred = clf.predict(test_data)
    report = classification_report(test_labels, y_pred)
    print("Classification Report for", clf.__class__.__name__)
    print(report)
    print()
```

```python
# c. Calculate Recall Precision and F1 score of all the models

for clf in classifiers:
    y_pred = clf.predict(test_data)
    report = classification_report(test_labels, y_pred, output_dict=True)
    recall = report['macro avg']['recall']
    precision = report['macro avg']['precision']
    f1_score = report['macro avg']['f1-score']
    print("Metrics for", clf.__class__.__name__)
    print("Recall:", recall)
    print("Precision:", precision)
    print("F1-Score:", f1_score)
    print()
```

```python
# d. Visualize confusion matrix using heatmaps

for clf in classifiers:
    y_pred = clf.predict(test_data)
    cm = confusion_matrix(test_labels, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap="Blues")
    plt.title("Confusion Matrix for " + clf.__class__.__name__)
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.show()

```

```python
# e. Select the best model based on the best accuracies

best_accuracy = 0
best_model = None

for clf in classifiers:
    y_pred = clf.predict(test_data)
    accuracy = accuracy_score(test_labels, y_pred)
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = clf

print("Best Model:", best_model.__class__.__name__)
print("Best Accuracy:", best_accuracy)
```
